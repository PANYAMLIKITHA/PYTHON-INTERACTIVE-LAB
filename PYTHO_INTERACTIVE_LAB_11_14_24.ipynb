```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

arr = [1, 3, 5, 7, 9, 11, 13]
target = 7
result = binary_search(arr, target)
print(f"Element found at index: {result}")
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(5))
```

Student Management System (OOP, File Handling)

```python
import os
import pickle

# Class to represent each student
class Student:
    def __init__(self, student_id, name, age, grade):
        self.student_id = student_id
        self.name = name
        self.age = age
        self.grade = grade

    def __str__(self):
        return f"ID: {self.student_id}, Name: {self.name}, Age: {self.age}, Grade: {self.grade}"

    def to_dict(self):
        return {"student_id": self.student_id, "name": self.name, "age": self.age, "grade": self.grade}

# Class to manage the student database
class StudentManager:
    def __init__(self, filename="students.db"):
        self.filename = filename
        self.students = self.load_data()

    def load_data(self):
        """Load student data from file."""
        if os.path.exists(self.filename):
            with open(self.filename, "rb") as file:
                return pickle.load(file)
        return []

    def save_data(self):
        """Save student data to file."""
        with open(self.filename, "wb") as file:
            pickle.dump(self.students, file)

    def add_student(self, student):
        """Add a student to the database."""
        self.students.append(student)
        self.save_data()

    def view_students(self):
        """Display all students."""
        if not self.students:
            print("No students found!")
        else:
            for student in self.students:
                print(student)

    def delete_student(self, student_id):
        """Delete a student by ID."""
        self.students = [student for student in self.students if student.student_id != student_id]
        self.save_data()

# Main program
def main():
    manager = StudentManager()

    while True:
        print("\nStudent Management System")
        print("1. Add Student")
        print("2. View Students")
        print("3. Delete Student")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            student_id = input("Enter Student ID: ")
            name = input("Enter Name: ")
            age = int(input("Enter Age: "))
            grade = input("Enter Grade: ")
            student = Student(student_id, name, age, grade)
            manager.add_student(student)
            print("Student added successfully!")

        elif choice == "2":
            manager.view_students()

        elif choice == "3":
            student_id = input("Enter Student ID to delete: ")
            manager.delete_student(student_id)
            print(f"Student with ID {student_id} deleted successfully!")

        elif choice == "4":
            print("Exiting...")
            break

        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with student grades
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Math': [85, 92, 78, 90, 88, 75, 95],
    'English': [78, 85, 90, 87, 92, 80, 91],
    'Science': [90, 88, 85, 91, 95, 84, 89]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Show the DataFrame
print("Student Grades:")
print(df)

# Plot the grades of each subject
df.plot(x='Name', y=['Math', 'English', 'Science'], kind='bar', figsize=(10,6))
plt.title('Student Grades in Math, English, and Science')
plt.xlabel('Students')
plt.ylabel('Grades')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Calculate average grades for each student
df['Average'] = df[['Math', 'English', 'Science']].mean(axis=1)

# Show the updated DataFrame with Average grades
print("\nUpdated Student Grades with Averages:")
print(df)

# Plot the average grades of each student
df.plot(x='Name', y='Average', kind='line', marker='o', color='red', figsize=(10,6))
plt.title('Average Grades of Students')
plt.xlabel('Students')
plt.ylabel('Average Grade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

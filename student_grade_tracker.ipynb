```python
# Function to calculate the letter grade
def get_letter_grade(average):
    if average >= 90:
        return 'A'
    elif average >= 80:
        return 'B'
    elif average >= 70:
        return 'C'
    elif average >= 60:
        return 'D'
    else:
        return 'F'

# Function to calculate the GPA
def calculate_gpa(average):
    if average >= 90:
        return 4.0
    elif average >= 80:
        return 3.0
    elif average >= 70:
        return 2.0
    elif average >= 60:
        return 1.0
    else:
        return 0.0

# Function to track and manage student grades
def track_grades():
    print("Welcome to the Student Grade Tracker!\n")

    student_name = input("Enter the student's name: ")
    subjects = int(input("Enter the number of subjects/assignments: "))

    grades = []

    # Input grades for each subject/assignment
    for i in range(subjects):
        subject_name = input(f"Enter the name of subject {i + 1}: ")
        grade = float(input(f"Enter the grade for {subject_name}: "))
        grades.append(grade)

    # Calculate the average grade
    average_grade = sum(grades) / len(grades)

    # Calculate letter grade and GPA
    letter_grade = get_letter_grade(average_grade)
    gpa = calculate_gpa(average_grade)

    # Display the results
    print("\n---- Student Grade Summary ----")
    print(f"Student Name: {student_name}")
    print(f"Average Grade: {average_grade:.2f}")
    print(f"Letter Grade: {letter_grade}")
    print(f"GPA: {gpa:.2f}")

#  program to run the grade tracker
if __name__ == "__main__":
    track_grades()
```

```python
import datetime

# Sample data structure for storing library items
library_items = []

# Sample data structure for storing checked-out items
checked_out_items = {}

# Fine per day for overdue items (in dollars)
fine_per_day = 1.00

# Item class to represent books, magazines, and DVDs
class LibraryItem:
    def __init__(self, title, category, author=None):
        self.title = title
        self.category = category  # Can be 'Book', 'Magazine', or 'DVD'
        self.author = author
        self.checked_out = False
        self.due_date = None

    def __str__(self):
        return f"{self.title} ({self.category}) by {self.author if self.author else 'N/A'}"

# Function to add a new library item
def add_item():
    title = input("Enter the title of the item: ")
    category = input("Enter the category (Book, Magazine, DVD): ").capitalize()
    author = None
    if category == 'Book':
        author = input("Enter the author: ")

    new_item = LibraryItem(title, category, author)
    library_items.append(new_item)
    print(f"Item '{title}' added successfully.\n")

# Function to search for items
def search_item():
    search_term = input("Enter the title, author, or category to search: ").lower()
    results = []
    for item in library_items:
        if search_term in item.title.lower() or search_term in (item.author or "").lower() or search_term in item.category.lower():
            results.append(item)

    if results:
        print("\nSearch Results:")
        for item in results:
            print(item)
    else:
        print("No items found.\n")

# Function to check out an item
def checkout_item():
    title = input("Enter the title of the item to check out: ")
    for item in library_items:
        if item.title.lower() == title.lower() and not item.checked_out:
            item.checked_out = True
            item.due_date = datetime.date.today() + datetime.timedelta(days=14)  # Due in 14 days
            checked_out_items[item.title] = item
            print(f"Item '{item.title}' checked out successfully. Due date: {item.due_date}\n")
            return
    print("Item not found or already checked out.\n")

# Function to return an item
def return_item():
    title = input("Enter the title of the item to return: ")
    for item in checked_out_items.values():
        if item.title.lower() == title.lower():
            item.checked_out = False
            overdue_days = (datetime.date.today() - item.due_date).days
            overdue_fine = max(0, overdue_days) * fine_per_day
            checked_out_items.pop(item.title)
            if overdue_fine > 0:
                print(f"Item returned. You have an overdue fine of ${overdue_fine:.2f} for {overdue_days} overdue days.\n")
            else:
                print("Item returned successfully. No overdue fine.\n")
            return
    print("Item not found or not checked out.\n")

# Function to view all library items and their status
def view_all_items():
    if library_items:
        print("\nLibrary Items:")
        for item in library_items:
            status = "Checked Out" if item.checked_out else "Available"
            print(f"{item} - {status}")
    else:
        print("No items in the library.\n")

# Menu function for the library system
def menu():
    while True:
        print("\nLibrary Management System")
        print("1. Add new item")
        print("2. Search for an item")
        print("3. Checkout an item")
        print("4. Return an item")
        print("5. View all items")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            add_item()
        elif choice == '2':
            search_item()
        elif choice == '3':
            checkout_item()
        elif choice == '4':
            return_item()
        elif choice == '5':
            view_all_items()
        elif choice == '6':
            print("Exiting the system.")
            break
        else:
            print("Invalid choice, please try again.")

# Run the system
menu()
```
